```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
filein = "../camb.z=47.ascii"
input_laggragic = pd.read_csv(filein, skiprows=5, delim_whitespace=True, names=["k/h","P_nonu","P_cdm","P_bar","f_nonu","f_cdm","f_bar","D_nonu","D_cdm","D_bar"])
```

```python
plt.loglog(input_laggragic["k/h"],input_laggragic["P_nonu"],label='input_power_spectrum')
```

```python
logk=np.log10(input_laggragic["k/h"])
Delta2=input_laggragic["P_nonu"]*(input_laggragic["k/h"]**3)/(2*np.pi**2)
logDelta2=np.log10(Delta2)
```

```python
plt.plot(logk,logDelta2)
```

```python
df = pd.DataFrame({'logk': logk,'logDelta2': logDelta2})
df.to_csv('z=47.txt', sep=' ', index=False, header=False)
```
